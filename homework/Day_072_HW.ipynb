# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def relu(x):
    array = list(map(lambda x: x if x>0 else 0, x))
    return array

def gradrelu(x):
    array = list(map(lambda x: 1 if x>0 else 0, x))
    return array

x = plt.linspace(-10, 10, 100)
```

```python
# ReLU function
plt.plot(x, relu(x), 'b')
plt.grid()
plt.title('ReLU function')
plt.text(-7, 8, r'$f(x)=max({0},{x})$', fontsize=15)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```

```python
# dReLU function
plt.plot(x, gradrelu(x), 'r')
plt.grid()
plt.title('dReLU function')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(.1))
```
